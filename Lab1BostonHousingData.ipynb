```python
#0 Basic Import
import os
from google.colab import drive
drive.mount('/content/gdrive')
os.listdir('/content/gdrive/MyDrive')
#Basic Packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
#Scikit-learn
import sklearn
from sklearn.cluster import KMeans
```

```python
#Exercise 2 Boston Housing Data
#a
url='https://raw.githubusercontent.com/MajdiRadaideh/S097data/main/boston.csv'
data= pd.read_csv(url)
#b
print("First 5: ", data.head(5))
print("Last 5: " ,data.tail(5))
#c
print("shape: ", data.shape)
#d
print("Col Names: ", data.columns)

#e
print("Col 3: ", data.iloc[:,2])

#f
print("Mean of Tax", data["TAX"].mean(axis=0))

#g
print("Standard Dev of TAX :", data["TAX"].std(axis=0))

#h
print("Max :", data["B"].max())
print("Min :", data["B"].min())

#i
print("Describe Data:\n" ,data.describe())
print("Describe Tax: \n",data["TAX"].describe())
print("Describe B: \n", data["B"].describe())

#j
idx = list(range(0,len(data)))
plt.plot(idx, data["LSTAT"],data["MEDV"], '-',label='NAME')


```

```python
#Exercise 3
X = data[["RM","LSTAT","MEDV"]]
km = KMeans(n_clusters = 3 , n_init=30 , random_state=1)
km.fit(X)
labels =km.labels_
#Plotting
fig = plt.figure(1, figsize=(7,7))
ax = Axes3D(fig, rect=[0, 0, 0.95, 1], elev=48, azim=134)
ax.scatter(X['RM'], X['LSTAT'], X['MEDV'], c=labels.astype(np.float), s=50)
ax.set_xlabel("RM")
ax.set_ylabel("LSTAT")
ax.set_zlabel("MEDV")
plt.title("K Means", fontsize=14)
plt.show()
```
